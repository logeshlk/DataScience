```python
test = str(input('Enter the input for reverse string :'))

print ('Entered input is :',test)

revstr = test[::1]

revstr1 = ''.join(reversed(revstr))

print ('Reverse String of Entered input :', revstr)
print ("Original Input :",revstr1)
```
